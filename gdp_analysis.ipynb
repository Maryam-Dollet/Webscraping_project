```python
host_city_df = pd.read_csv("data_csv/host_city_positions.csv", sep=";")
```

```python
host_city_df
```

On filtre que à partir de l'année 1960 car les données du PIB commencent à partir de 1960.

```python
df1 = host_city_df[host_city_df["Annee"] >= 1960]
df1
```

```python
df1 = df1.rename(columns={"country": "iso2"})
```

```python
df1["Game"] = df1["Lieu"].astype(str) + " " + df1["Annee"].astype(str)
```

```python
df1
```

```python
df1.head(1)[["Game", "iso2", "Annee"]].merge(gdp_countries, how="left", on="iso2")
```

```python
df1.iloc[[0]]
```

```python
for i in range(len(df1)):
    print(df1.iloc[[i]][["Game", "iso2", "Annee"]].merge(gdp_countries, how="left", on="iso2"))
```

```python
athene_df = df1.iloc[[4]][["Game", "iso2", "Annee"]].merge(gdp_countries, how="left", on="iso2").melt(id_vars=["Game", "Annee","country_name", "iso2"], var_name="Year", value_name="GDP")
```

```python
athene_df
```

```python
athene_df.Year =  athene_df.Year.astype(int)
```

```python
athene_df.info()
```

```python
test_year = athene_df.iloc[0].Annee
```

```python
df2 = athene_df[athene_df["Year"].between(test_year, test_year + 7)]
```

```python
df2
```

```python
fig = px.line(df2, x="Year", y="GDP")
fig
```
