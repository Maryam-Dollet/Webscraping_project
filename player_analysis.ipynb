Rename the columns

```python
events.columns
```

```python
events["medal"] = events["medal"].fillna(np.nan)
```

```python
events["medal"]
```

```python
events.medal.value_counts()
```

```python
events["year"] = events["game"].apply(lambda x: str(x).split()[0])
```

```python
events
```

```python
events["season"] = events["game"].apply(lambda x: " ".join(str(x).split()[1:-1]) if len(str(x).split()[1:]) > 0 else np.nan)
```
